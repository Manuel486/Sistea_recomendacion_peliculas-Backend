```python
# Configura la conexión a MySQL
conexion = mysql.connector.connect(
    host="monorail.proxy.rlwy.net",
    port="19352",
    user="root",
    password="2a6HF-cG6C3e-1Bg3eC-F4Bde6hcAh4g",
    database="srp",
)

# Ruta al archivo CSV que deseas cargar
archivo_csv = "./data/links_data.csv"

try:
    # Lee el archivo CSV en un DataFrame
    df = pd.read_csv(archivo_csv)

    # Convierte los nombres de las columnas a minúsculas
    df.columns = [col.lower() for col in df.columns]

    # Carga los datos en la tabla existente en MySQL
    cursor = conexion.cursor()

    for index, row in df.iterrows():
        imdb_id = int(row["imdbid"])
        tmdb_id = int(row["tmdbid"])
        movie_id = int(row["movieid"])

        # Verifica si el movieId existe en la tabla movie
        cursor.execute("SELECT COUNT(*) FROM movie WHERE id = %s", (movie_id,))
        count = cursor.fetchone()[0]

        if count > 0:
            # El movieId existe, se puede insertar en la tabla link
            cursor.execute(
                """
                INSERT INTO link (movieId, imdbId, tmdbId)
                VALUES (%s, %s, %s)
                """,
                (movie_id, imdb_id, tmdb_id),
            )
        else:
            print(f"El movieId {movie_id} no existe en la tabla movie. Fila ignorada.")


    conexion.commit()

    print("Datos cargados exitosamente en la tabla de MySQL.")

except mysql.connector.Error as err:
    print("Error al cargar datos en la tabla de MySQL:", err)

finally:
    # Cierra el cursor y la conexión a MySQL
    cursor.close()
```
